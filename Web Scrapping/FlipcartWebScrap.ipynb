# Web Scrapping

```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import logging
```

```python
flipcart_url = "https://www.flipkart.com/search?q="+"iphone12"
```

```python
flipcart_url
```

```python
urlclient = uReq(flipcart_url)
urlclient
```

```python
flipcart_page = urlclient.read()
flipcart_page
```

```python
flipcart_html = bs(flipcart_page, 'html.parser') # Used to align the html page or beautfy the html page.
```

```python
flipcart_html
```

```python
big_box = flipcart_html.findAll('div', {"class":"lvJbLV col-12-12"}) # It find the each division having the class passed to it(each i[phone on the page) 
```

```python
len(big_box)
```

```python
del big_box[0:3] # It delets the first 3 boxes beacause it contains the vav bar and other page details.
```

```python
product_link = "https://www.flipkart.com"+big_box[5].div.div.div.a['href']
```

```python
product_req = requests.get(product_link) 
```

```python
product_link
```

```python
product_html = bs(product_req.text, 'html.parser')
```

```python
comment_box = product_html.find_all("div", {"class":"mlW33x"})
comment_box
```

```python
len(comment_box)
```

```python
# Printing the names of each reveiwer who posted their review
for i in comment_box:
    print(i.div.div.find_all('p',{"class":"zJ1ZGa ZDi3w2"})[0].text)
```

```python
# Finding the ratings of each reviewer
for i in comment_box:
    print(i.div.div.find_all('div',{"class":"MKiFS6 ojKpP6"})[0].text)


# Or We Can Also find this by using another way

# for i in comment_box:
#     print(i.div.div.div.div.text)
```

```python
# Finding all the Comment/reveiw Headers of each reveiw
for i in comment_box:
    print(i.div.div.div.p.text)
```

```python
# Finding the comments of each reveiw
for i in comment_box:
    print(i.div.div.find_all('div',{"class":""})[0].text)
```
