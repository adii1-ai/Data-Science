## Opening CSV file from URL

```python
import requests
from io import StringIO

url="https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

## parameters to read_csv()

### 1.  sep parameter

```python
# sep parameter is used to separate the raw data and make dataframe
# if file is csv(comma separated value) then sep =','
# if file is tsv(tab separated value) then sep ='\t'
```

```python
pd.read_csv("movie_titles_metadata.tsv") # Without sep parameter
```

```python
pd.read_csv("movie_titles_metadata.tsv", sep = '\t')

# In this dataframe first row becomes the column name thus use name parameter 
```

### 2. name parameter

```python
# used to set the name to the each column
```

```python
pd.read_csv("movie_titles_metadata.tsv", sep = '\t',names=["sno","name","release_year","rating","votes","genres"])
```

### 3. index_col parameter

```python
# index_col is used to make the index column from the dataframe insteed of making separate index column
```

```python
pd.read_csv("aug_train.csv")
```

```python
pd.read_csv("aug_train.csv", index_col='enrollee_id')
```

### 4. header parameter

```python
# without header when we read csv file then column names inside that csv file are are considered as data and added inside dataframe.
```

```python
pd.read_csv('test.csv') # Without header headers isdide a raw data are are treated as row so table headers are not printed inside headr row.
```

```python
pd.read_csv('test.csv', header = 1) # when we set header=1 then the header ( i.e column names ) are displayed as column name insteed data
```

### 5. usecols parameter 

```python
# When we want to use particular columns then we use parameter use_cols
# Some times it is neccessary to use particular columns only then this parameter is used.
```

```python
pd.read_csv('aug_train.csv') # without use_cols
```

```python
pd.read_csv('aug_train.csv', usecols = ['enrollee_id','city','gender'])
```

### 6. skiprows/nrows parameter

```python
# We pass the list of row indices that we want to skip or not want to add in the dataframe.
# when the dataset is very big and system does not have that much storage to load dataset then skiprows parameter is used
```

```python
pd.read_csv('aug_train.csv')
```

```python
pd.read_csv('aug_train.csv', skiprows=[0,1,2,3])
```

### 7. Encoding parameter

```python
# When we try to read file and its encodingtype (UTF-8--> by Default) is different then we have to change the its encoding type.
```

```python
pd.read_csv('Zomato-data-.csv',encoding='latin-1') 
# In this example there is no issue with encoding.
```

### 8. on_bad_lines parameter

```python
# on_bad_lines parameter is used to skip the lines which includes more number of columns of data 
# if the number of columns of data is more than other rows it affects the dataframe thus we have to skip those lines
```

```python
pd.read_csv('test.csv',header=1,index_col=0, on_bad_lines='skip')
```

### 9. dtype parameter

```python
# dtype is used to convert the column of one data type into another data type 
# We have to pass the dictionary of key value where key is the column name and value is the data type in which we want to convert the data
# example., we have to convert the column containing float value into integer then we use dtype parameter.
```

```python
train = pd.read_csv('aug_train.csv') # in this dataset target column has float datatype
train['target'].dtype
```

```python
train.info() # checking the information of each column
```

```python
# Now we use the dtype parameter and convert the target column (float) into integer
pd.read_csv('aug_train.csv', dtype={'target':int})
```

### 10. Handling Dates

```python
df1 = pd.read_csv('IPL Matches 2008-2020.csv')
df1['date'].dtype   # here dtye is O means Object(i.e String)
```

```python
pd.read_csv('IPL Matches 2008-2020.csv', parse_dates=['date']) # here date type (String) is converted into date for using its functionality
```

### 11. Convertors

```python
def rename(name):
    if(name == 'NaN'):
        return '------'
    else:
        return name
```

```python
pd.read_csv('aug_train.csv')
```

```python
# Conversion
# - Used to change the value of a columns by writing the function to it and passing the function name as converters parameter


pd.read_csv('aug_train.csv',converters={'company_type':rename}) # here rename function is called to rename the column values
```

### 12. na_values parameter

```python
pd.read_csv('aug_train.csv')
```

```python
pd.read_csv('aug_train.csv',na_values = ['Male']) # Here Make data values are converted into NaN values.
```

### 13. Loading a huge dataset in Chunks

```python
# When the dataset is very large/Big and we perform any operation then due to less memory it not works.
# Thus while loading we load this Big Dataset inside Chunks of Same size.
# parameter -->> chunksize = size_of_chunk
```

```python
df = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunks in df:
    print(chunks.shape)
```
