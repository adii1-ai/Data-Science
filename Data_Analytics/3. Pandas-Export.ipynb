```python
import pandas as pd
```

## 1. to_csv

```python
df = pd.read_csv('deliveries.csv')
df
```

```python
temp_df = df.groupby('driver_id')['delivery_distance_meters'].sum().reset_index()

# here sum of delivey_distannce in meters of each driver is printed

temp_df
```

```python
temp_df.to_csv('drivers_total_distance.csv',index = False)  # In the current folder csv file will be created.
```

```python
temp_df
```

## 2. to_excel()

```python
temp_df.to_excel('drivers_total_distance.xlsx')  # excel file will created in current folder
```

### giving Excel sheet name

```python
temp_df.to_excel('output.xlsx',sheet_name='drivers')
```
