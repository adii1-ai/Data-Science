## 1. EDA
- Exploratory Data Analysis (EDA) is the process of analyzing datasets to summarize their main characteristics, discover patterns, detect anomalies, test assumptions, and check data quality using statistics and visualizations before applying formal models.

## 2. Objectives of EDA
- Understand dataset structure and variables
- Identify missing values and duplicates
- Detect outliers and anomalies
- Study data distributions
- Analyze relationships between variables
- Decide preprocessing and feature engineering steps
## 3. Types of EDA
### 3.1 Univariate Analysis
Analysis of one variable at a time.

`Numerical Variables`
- Mean, Median, Mode
- Variance, Standard Deviation
- Min, Max, Quartiles

`Categorical Variables`
- Frequency count
- Percentage distribution

`Common Plots`

- Histogram
- Box Plot
- Bar Chart

### 3.2 Bivariate Analysis
Analysis of two variables together.

`Numerical vs Numerical`
- common method -> Correlation
- plot -> scatter plot(regression plot), 2D histplot, 2D KDE plots

`Numerical vs Categorical`
- common method -> Group statistics
- plot -> Box plot, kde plot, barplot, violineplot, scatterplot

`Categorical vs Categorical`
- common method -> Cross‑tabulation
- plot -> Heatmap, stacked barplots, treemaps

### 3.3 Multivariate Analysis
Analysis involving more than two variables.
- Pair Plot
- Correlation Matrix
- Heatmap
- Dimensionality reduction (PCA – basic understanding)

## 4. Steps in EDA
Raw Data -> Understand Variables -> Data Cleaning -> Univariate Analysis -> Bivariate Analysis -> Multivariate Analysis -> Insights & Decisions

## 5. Data Cleaning in EDA

- Handling missing values (mean, median, mode, drop)
- Removing duplicates
- Fixing incorrect data types
- Handling inconsistent units (e.g., "10u", "20mg")
- Renaming columns

## 6. Outlier Detection

`Methods`:

- IQR Method
- Z‑Score Method
- Box Plot visualization

Why important? 
 -> Outliers can heavily affect mean, variance, and ML models.

## 7. Importance of EDA

- Prevents wrong assumptions
- Improves model accuracy
- Detects data leakage
- Helps in feature selection
- Saves time in later stages

`"A model is only as good as the data behind it."`
