## Categorical Distribution Plot

```python
# It gives Univariate Analysis of data
```

### boxplot()

```python
# Draw a box plot to show distributions with respect to categories.
# boxplot is a standardized way of displaying the distribution of data based on a five number summary.
# (minimum, first quartile(Q1), median, third quartile(Q3), and maximum).
# boxplots can tell you about your outliers and what their values are.
# boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

# Values less than the minimun and greater than the maximum is called the Outliers
```

```python
sns.boxplot(data=tips, x='sex', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
#  using catplot
sns.catplot(data=tips, x = 'day', y='total_bill', hue='sex', kind='box')
```

```python
# plot of total bill

sns.boxplot(data=tips, y='total_bill')
```

### violinplot()

```python
# Violinplot = Boxplot + KDE Plot
```

```python
sns.violinplot(data=tips, x = 'day', y='total_bill')
```

```python
# using catplot
sns.catplot(data=tips, x = 'day', y='total_bill', hue='sex', kind='violin')
```

```python
# split=True parameter

sns.catplot(data=tips, x = 'day', y='total_bill', hue='sex', kind='violin', split=True)
```

### boxenplot()

```python
# Draw an enhanced box plot for larger datasets.
# This style of plot was originally named a “letter value” plot because it shows a large number of quantiles that are defined as “letter values”.
# It is similar to a box plot in plotting a nonparametric representation of a distribution in which all features correspond to actual observations.
# By plotting more quantiles, it provides more information about the shape of the distribution, particularly in the tails.
```

```python
sns.boxenplot(data=tips, x = 'day', y='total_bill')
```

## Categorical Estimate Plot

### barplot()

```python
# when there are multiple observations in each category, it uses bootstraping to compute a confidence interval around the estimate, which is plotted using error bars
```

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
# errorbar=()
    # -errorbar=('ci', width)  --ci = Confidence Interval
    # -errorbar('pi', width)   --pi = Percentile Interval
    # -errorbar('sd', scale)   --sd = Standard Deviation
    # -errorplot('se', scale)  --se = Standard Error

sns.barplot(data=tips, x='sex', y='total_bill',errorbar=('pi', 100))
```

```python
# hue parameter

sns.barplot(data=tips, x='sex', y='total_bill',errorbar=('ci', 100),hue='smoker')
```

```python
# estimate parameter

import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker', estimator = np.min)

# In above example estimator parameter is used to plot the minimun total_bill of smoker male , Non-smoker male and Smoker-Female and Non-Smoker Female 
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker', estimator = np.max)

# estimator = np.max ==>> plots the maximum value of total_bill of smoker male , Non-smoker male , Smoker Female and Non-smoker Female  
```

### pointplot()

```python
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker', estimator = np.min)
```

```python
# Male Smoker, Male Non-Smoker, Female Smoker and Female Non-Smoker

sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker')
```

```python
# errorbar = None  ==>> plots the graph without errorbar
# barplot Focuses on Comparison

sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker', errorbar= None)
```

### countplot()

```python
# shows number of males and females on each day
sns.countplot(data=tips, x='sex', hue='day')
```

```python
# faceting using catplot

sns.catplot(data=tips,x='sex',y='total_bill',row='day',col='smoker', hue='sex', kind='box')

# This plots the graphs of each day row wise for smoker and Non Smoker in column wise
```
