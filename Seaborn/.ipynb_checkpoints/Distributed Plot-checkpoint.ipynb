# Distributed Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

## histplot()

```python
# It is a axes level function
```

```python
# Plotting Univariate histogram

sns.histplot(data = tips, x = 'total_bill')
```

## displot()

```python
sns.displot(data = tips, x = 'total_bill', kind = 'hist')
```

### bins Parameter

```python
sns.displot(data = tips, x = 'total_bill', kind = 'hist', bins = 4)
```

```python
# It is also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

tips
```

```python
# we can say it as countplot because we plotted the graph based on categorical data

# countplot
sns.displot(data = tips, x = 'day', kind = 'hist', hue='sex')
```

### hue parameter

```python
sns.displot(data = tips, x = 'tip', hue = 'sex')
```

### element parameter

```python
# It is used to visualize the chart better
# It only creates the outlines of the bar makes visualization easy 
# due to hue parameter visualization is slightly difficult(not understandable)

# Syntax:-
# element = 'step'
```

```python
sns.displot(data = tips, x = 'tip', hue = 'sex',kind = 'hist', element = 'step')
```

### row and col parameter

```python
# Adding row or col parameter is called faceting/ Facetin using the row and col
# Faceting only work on figure level functions
# Faceting not work on histplot function 

# It is used to plot the charts based on the features/column of dataset
# For each different value inside the feature/column passed to the row or col parameter grapg is plotted.
```

```python
sns.displot(data = tips, x = 'tip', hue = 'sex', row = 'sex', col = 'smoker')
```

```python
# hue parameter 
titanic = sns.load_dataset('titanic')
titanic
```

```python
# hue 
sns.displot(data = titanic, x = 'age', hue = 'sex', kind = 'hist')
```

```python
# element='step'

sns.displot(data=titanic, x='age', kind='hist',hue='sex', element='step')
```

## kdeplot()

```python
# kde  stands for kernel density estimate (KDE) 
# Univariate/ Bivariate 
# kdeplot() is axes level function

# Basically it shows that the Probability/Density of the values on x axis
```

```python
tips.head(20)
```

```python
sns.kdeplot(data=tips, x='total_bill')

# In this graph the probability of the bill is showed
```

```python
# Plotting the kdeplot using displot

sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
# hue

sns.kdeplot(data=tips,x='total_bill',hue='sex')

# This plot shows the probability of the total bill of Male and Female
```

```python
# fill=true 
# -It fills the plot 

sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

## rugplot()

```python
# -Plot marginal distributions by drawing ticks along the x and y axes.
# -
```

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

## Bivariate histplot

```python
# Bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observation within each rectangle with the fill color
```

```python
sns.histplot(data=tips, x='total_bill',y='tip')
```

```python
# using displot
sns.displot(data=tips, x='total_bill',y='tip',kind='hist')
```

```python
# Bivariate kdeplot

# -a bivariate KDE plot smoothes the (x,y) observation with a 2D gaussian
```

```python
sns.kdeplot(data=tips, x='total_bill',y='tip')
```
