# Regression Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

## regplot()

```python
sns.regplot(data=tips, x='total_bill', y='tip')
# we cannot add hue parameter in regplot/(Axix level function)

# In this graph the middle line is said as regression line 
# On the basis of this line ml model do the prediction 
# blue color space around the line shows the error in the prediction of model
```

## lmplot()

```python
# Using lmplot()
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
# hue parameter
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

## residplot()

```python
sns.residplot(data=tips, x='total_bill', y='tip')
```
