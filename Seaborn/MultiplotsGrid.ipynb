# Multiplots Grid

```python
Plots Under Multiplots grids
    -Facet grids
        -FacetGrid
    -Pair grids
        -pairplot
        -pairGrid
    -Joint grids
        -jointplot
        -jointGrid
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

# FacetGrid()

```python
# Multi-plot grid for plotting conditional relationships.
```

```python
g = sns.FacetGrid(data=tips,row='time', col='day', hue='sex')
g.map(sns.scatterplot,'total_bill', 'tip')
```

```python
g = sns.FacetGrid(data=tips,row='time', col='day', hue='sex')
g.map(sns.histplot,'total_bill')
```

## pairplot()

```python
# Plot pairwise relationships in a dataset.
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
# Automatically detects the numerical columns from dataset and plot the pairwise plot

sns.pairplot(iris)

# If both the columns are same then it plots the histogram
```

```python
#  hue parameter
sns.pairplot(iris, hue='species')
```

## PairGrid()

```python
# It provides the better flexibility over pairplot()
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(sns.histplot)       # at diagonal it plots the histogram plots
g.map_offdiag(sns.scatterplot) # at off diagonal it plots the scatterplots
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.violinplot)     # at diagonal it plots the violinplot
g.map_offdiag(sns.scatterplot) # at off diagonal it plots the scatterplots
```

```python
# map_diag() -> map_upper() -> map_lower()

g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)     # at diagonal it plots the histogram
g.map_upper(sns.kdeplot)     # at upper diagonal it plots the kdeplot
g.map_lower(sns.scatterplot) # at lower diagonal it plots the histogram
```

```python
#  vars=[] parameter ==>> It includes the column only on which the plots are plotted

g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)     # at diagonal it plots the histogram
g.map_upper(sns.kdeplot)     # at upper diagonal it plots the kdeplot
g.map_lower(sns.scatterplot) # at lower diagonal it plots the histogram
```

## jointplot()

```python
sns.jointplot(data=tips, x='total_bill', y='tip')

# by default It plots scatterplot and also the histogram on indivisual columns
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip',kind='kde')

# using kind parameter
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip',kind='reg')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip',kind='resid')
```

## JointGrid()

```python
g = sns.JointGrid(data=tips,x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.histplot) 
# this is the flexibility provided by the JointGrid() over the jointplot()
# Means we can choose/ plot the graph we want
# In jointplot() there is limited options of plots to plot the graph
```

```python
g = sns.JointGrid(data=tips,x='total_bill', y='tip')
g.plot(sns.kdeplot, sns.histplot) 
```

```python
g = sns.JointGrid(data=tips,x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.violinplot) 
```

```python
g = sns.JointGrid(data=tips,x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.violinplot) 
```
