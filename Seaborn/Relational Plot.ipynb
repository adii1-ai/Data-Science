# Relatonal Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

## relplot()

```python
sns.relplot(data =tips, x ='total_bill', y = 'tip', kind = 'scatter') 
# We can print the line plot by using the relplot() function by just changing the kind = 'line'
```

## scatterplot()

```python
sns.scatterplot(data = tips, x ='total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size')
```

## lineplot()

```python
# lineplot is a axes level function 
```

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# Plotting the graph
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# Using relplot()

sns.relplot(data=temp_df, x='year', y='lifeExp', kind = 'line')
```

```python
# Taking only Three country's data 
# hue parameter

temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', kind = 'line', hue = 'country')
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue = 'country')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', kind = 'line', hue = 'country', style = 'continent', size='continent')
```

## facet plot

```python
tips
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'line', col = 'smoker', row = 'sex')
```

```python
# col and col_wrap parameters
gap
```

```python
# col parameter decides the plotting of graphs in columns based on value passed to it.

sns.relplot(data = gap, x = 'lifeExp', y = 'gdpPercap', kind  = 'scatter', col = 'year') 
```

```python
#  col_wrap parameter desides the how many columns of graph should be plloted

sns.relplot(data = gap, x = 'lifeExp', y = 'gdpPercap', kind  = 'scatter', col = 'year', col_wrap = 3) 
```

```python
# plotting separate graphs based on continents

sns.relplot(data = gap, x = 'lifeExp', y = 'gdpPercap', kind  = 'scatter', col = 'continent', col_wrap = 3) 
```
