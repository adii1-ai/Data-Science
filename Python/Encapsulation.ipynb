# Encapsulation

```python
# Binding of data and methods inside a class is called the Encapsulation
```

```python
# In this code modification of variables/ methods is allowed outside the class,
# But Inheritance does not allows the modification outside the class

class TestClass():
    def __init__(self):
        print("Inside Tast Class")
        self.city = "Pune"
        self.state = "Maharashtra"

    def display(self):
        print("City name is : ",self.city)
        print("State name is : ",self.state)

obj = TestClass()
obj.city = "Banglore"   # Updating the value of city variable 
obj.state = "karnataka"   # Updating the value of state variable

obj.display()    # only updated values are printed
```

```python
# So, in this example we implemented the Encapsulation

# __ (2 underscore) is used to make variable private which are not accessible outside the class


class TestClass():
    __class_var1 = 100  # Private Class Variable
    class_var2 = 500   # Public Class Variable
    
    def __init__(self):
        print("Inside Test Class")
        x = 10                       # local variable
        self.__city = "Pune"         # Private instance variable
        self.__state = "Maharashtra" # Private instance variable 
        self.country = "India"       # Public instance variable 

    def display(self):
        print("City name is : ",self.__city)
        print("State name is : ",self.__state)
        print("Country name is :",self.country)
        print("Class variable 1 is :", self.__class_var1)
        print("Class variable 2 is :", self.class_var2)

obj = TestClass()
obj.__city = "Banglore"   # Updating the value of city variable 
obj.__state = "karnataka"   # Updating the value of state variable
obj.country = "Bharat"    # Will be printed in the output

obj.display()               # Updated value outside the class is not printed 

obj.__class_var1 = 50000
obj.class_var2 = 60000

obj.display()
```
