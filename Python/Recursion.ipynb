```python
def printNum(arr, sI):
    # Base Condition
    if(sI >= len(arr)):
        return

    # logic
    print(arr[sI])

    
    # Recursive call
    printNum(arr, sI+1)

printNum([1,3,4,5,7,8], 0)
```

```python
# print the numbers given in array in reverse order

def printRev(arr, sI):
    if(sI >= len(arr)):
        return

    printRev(arr, sI+1)

    print(arr[sI])

printRev([3,4,5,6,8], 0)
```

```python
def printNums(arr, sI):
    if(sI>=len(arr)):
        return

    print(arr[sI])
    printNums(arr, sI+1)
    print(arr[sI])
printNums([1,2,3,4,5],0)
```
