```python
class Mobile():
    x = 'python'
    def color(self):
        print('Black')
    def storage(self):
        print("8 GB RAM and 128 GB Storage")
vivo = Mobile()
vivo.color()
vivo.storage()
```

```python
class Main():
    x = 10
    y = 20
    def __init__(self):      # It is a Constructor, Executes when object of class is created
        print("value of x is ", self.x)   # self.x is used because x is class variable not a local variable(variable present inside a function

    def name(self):
        print("Value of y is ",self.y)
obj = Main()
obj.name() # calls the name function
```

```python
# Creating a instance variable -->>

class ClassName():
    x = 100 # class variable 

    def __init__(self):
        print("Inside a Cnstructor")
        self.y = 200   # Instance variable

    def Myfun(self):
        z = 300    # Local variable 
        print("Value of class variable x is ",self.x)
        print("Value of instance variable y is ",self.y)
        print("Value of local variable z is ", z)
obj = ClassName()
obj.Myfun()
```

## Passing Arguments

```python
class ClassName():
    x = 100
    def __init__(self, arg1, arg2):
        self.y = 200
        print("Inside a Constructor")
        print(arg1, arg2)

        self.argument_1 = arg1
        self.argument_2 = arg2

    def MyFun_1(self, arg3):
        
        global a
        a = 1500 # global variable
        
        print("Argument inside a function_1 is ", arg3)

        print(self.x, self.y)
        # We can access the arguments p[assed to another function  or constructor
        print("Arguments passed to the constructor " , self.argument_1 ,",",self.argument_2)

        
    def MyFun_2(self):
        print("Inside function 2")
        print(self.argument_1, self.argument_2)

        print("Value of a is ",a)

        # calling the function inside a function
        self.MyFun_1(300)

obj = ClassName(500, 1000)
obj.MyFun_1(900)
obj.MyFun_2()
```
