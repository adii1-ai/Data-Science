```python
class Calculation():
    def __init__(self, a = 10, b = 20):
        print(f'a == {a}')
        print(f'b == {b}')
        self.a = a
        self.b = b
    
    def addition(self):
        self.addition = self.a + self.b
        print(f'Addition of {self.a} and {self.b} is {self.addition}')

    def multiplication(self):
        mul = self.a * self.b
        print(f'Multiplication of {self.a} and {self.b} is {mul}')
        print("Addition  == ",self.addition)

c1 = Calculation(5,6)
c1.addition()
c1.multiplication()
```

## OOP Concepts

## Inheritance

```python
class company_info():
    def __init__(self, company_name, company_location):
        self.company_name = company_name
        self.company_location = company_location
        print("Inside Parent class constructor")

    def company_branch(self):
        branch_location = "Pune"
        print("Company name is ", self.company_name)
        print("Company location is ", self.company_location)
        print("Branch Location is " , branch_location)

class employee(company_info):
    def __init__(self, emp_name, emp_salary):
        self.emp_name = emp_name
        self.emp_salary = emp_salary
        print("Inside child class constructor")

        # Calling the parent class constructor
        company_info.__init__(self, "CloudInfoTech", "Kothrud") # Need to pass self keyword when calling the constructor using parent calss name

    def emp_info(self):
        print("Employee name is " , self.emp_name)
        print("Employee salary is ", self.emp_salary)
        
obj = employee("Karan", 50000)
obj.company_branch()
obj.emp_info()
```

```python
class company_info():
    def __init__(self, company_name, company_location):
        self.company_name = company_name
        self.company_location = company_location
        print("Inside Parent class constructor")

    def company_branch(self):
        branch_location = "Pune"
        print("Company name is ", self.company_name)
        print("Company location is ", self.company_location)
        print("Branch Location is " , branch_location)

class employee(company_info):
    def __init__(self, emp_name, emp_salary):
        self.emp_name = emp_name
        self.emp_salary = emp_salary
        print("Inside child class constructor")
        
        # Calling the parent class Constructor
        super().__init__("CloudInfoTech", "Kothrud") # When super() method is used while calling parent class constructor then noo need to pass self keyword

    def emp_info(self):
        print("Employee name is " , self.emp_name)
        print("Employee salary is ", self.emp_salary)
        
obj = employee("Karan", 50000)
obj.company_branch()
obj.emp_info()
```

## Two Ways to call the Parent Class constructor

## Types of Inheritance

## Single Inheritance

```python
class ParentClass():
    x = 10
    def parentMethod(self):
        print("Value of x is ", self.x)
        
class DerivedClass(ParentClass):
    y = 20
    def derivedMethod(self):
        print("Value of y is ", self.y)
        
obj = DerivedClass()
obj.parentMethod()
obj.derivedMethod()
```

## Multiple Inheritance

```python
class Parent_1():
    x = 10
    def Test_1(self):
        print("Parent Class 1")
        
class Parent_2():
    y = 20
    def Test_2(self):
        print("Parent Class 2")

class Derived2(Parent_1, Parent_2):
    def Test_3(self):
        print("Child Class")
D = Derived2()
D.Test_1()
D.Test_2()
D.Test_3()
```

## Multi Level Inheritance

```python
class Base():
    def parentMethod(self):
        print("Inside Parent class Method")

class Derived_1(Base):
    def derived_method1(self):
        print("Inside derived class 1 method")

class Derived_2(Derived_1):
    def derived_method2(self):
        print("Inside derived class 2 method")

D = Derived_2()
D.parentMethod()
D.derived_method1()
D.derived_method2()
```

## Hybrid Inheritance

```python
class Parent():
    def pclass_method(self):
        print("Inside the parent class method")

class child_1(Parent):
    def c_method(self):
        print("child 1 method")

class child_2(Parent):
    def c_function(self):
        print("child 2 method")

class grand_child(child_1, child_2):
    def g_method(self):
        print("grand child method")

D = grand_child()
D.pclass_method()
D.c_method()
D.c_function()
D.g_method()
```

## Hierarchical Inheritance

```python
class Parent():
    def p_method(self):
        print("parent class method")
class child_1(Parent):
    def c_method1(self):
        print("Child class 1 method")
class child_2(Parent):
    def c_method2(self):
        print("Child class 2 method")
D = child_1()
D1 = child_2()

D.p_method()
D.c_method1()

D1.p_method()
D1.c_method2()
```
