# Column Transformation:
- OneHot Encoding
- Ordinal Encoding

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer       # Used to impute the null values with mean, mode, median
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns =['has_covid']), df['has_covid'],test_size=0.2)
```

## Manually Transforming (Aam Zindagi)
- It requires more time to transform the columns if number of columns are more/large.

```python
# imputing the values using SimpleImputer()

si = SimpleImputer()

X_train_fever = si.fit_transform(X_train[['fever']])
X_test_fever = si.fit_transform(X_test[['fever']])
X_train_fever.shape
```

```python
# ordinal Encoding --> cough

oe = OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']])

X_test_cough = oe.fit_transform(X_test[['cough']])

X_train_cough.shape
```

```python
# OneHot Encoding --> gender, city

ohe = OneHotEncoder(drop = 'first', sparse_output = False)

X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])
X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])

X_train_gender_city.shape
```

```python
# Extracting Age

X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```python
# Merging the train data columns and test data columns

X_train_transformed = np.concatenate((X_train_age, X_train_fever, X_train_gender_city,X_train_cough),axis=1)
X_test_transformed = np.concatenate((X_test_age, X_test_fever, X_test_gender_city, X_test_cough), axis=1)

X_train_transformed.shape
```

## Mentos Zindagi 
-  Requires Less time Compared to above

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories = [['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(sparse_output = False, drop='first'),['gender','city'])
], remainder = 'passthrough')
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_train).shape
```
