- frequent value imputation is also called as a Mode Imputation.
- In this the most repeated value is imputed at missing place.
- Used when data is categorical.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gender = np.random.choice(['Male', 'Female'], size = 20)
gender
```

```python
city = np.random.choice(['nagar', 'Mandave'], size = 20)
city
```

```python
data = {
    'Gender' : ['Male', None, 'Female', None, 'Female', 'Female', 'Female',
       'Male', 'Female', 'Male', 'Female', 'Female', None, 'Male',
       'Female', 'Male', 'Male', 'Female', 'Female', 'Female'],
    'City' : [None, 'Mandave', 'Mandave', 'Mandave', 'nagar', 'nagar',
       None, 'Mandave', 'Mandave', 'Mandave', 'Mandave', 'Mandave',
       'Mandave', 'Mandave', 'nagar', 'Mandave', 'Mandave', 'nagar',
       'nagar', None]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
df['Gender'].value_counts() # Most frequent is Female
```

```python
df['City'].value_counts() # most frequent is Mandave
```

```python
df['Gender'].value_counts().plot(kind = "kde")
```

```python
df['City'].value_counts().plot(kind = 'kde')
```

```python
df['Gender'].value_counts().plot.bar()
```

```python
df['City'].value_counts().plot.bar()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
df[['Gender']] = imputer.fit_transform(df[['Gender']])
```

```python
df[['City']] = imputer.fit_transform(df[['City']])
```

```python
df
```

```python
df['Gender'].value_counts().plot(kind ='kde')
```

```python
df['City'].value_counts().plot(kind ='kde')
```
