```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'name' : ['Aditya','Karan','Samadhan','Babaji', 'Yash','Raj','Rushi','Keshav','Joy','Roy'],
    'Age' : [12,23, None ,44 ,None, 56,64,74,93,92],
    'Marks' : [20,40,20,49, None, 45,37,22,44,66]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Checking how many percantages of data is missing in each colummn 
df.isnull().mean()*100
```

```python
# If  less than equal to 5% data is missing then only apply Complete Case Analysis.
```

- Here Dataset is small, which is only for practise only so dont care about percatage of missing data

```python
# Apply an Complete Case Analysis
# Means remove the rows in which data is missing
```

```python
df.shape
```

```python
cols = df.columns
```

```python
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()
print(df.shape)
new_df.shape
```

```python
new_df.hist(bins = 10, density = True, figsize = (10,5))
plt.show()
```

```python
df.hist(bins=10, density = True, figsize = (10,5))
plt.show()
```

```python
# Checking performed CCA is right or wrong 
# If distribution is comparable then it s ok , If not comparable then not ok 

fig = plt.figure()
ax = fig.add_subplot()

# old column
df['Age'].hist(bins = 10 , ax = ax, density = True, color = 'Red')

# New column
new_df['Age'].hist(bins = 10, ax = ax, density = True, color = 'green', alpha = 0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
df['Marks'].hist(bins = 10, ax = ax, density= True, color = 'red')
new_df['Marks'].hist(bins = 10, ax = ax, density= True, color = 'green', alpha = 0.8)
```

```python
# both columns distribution before and after performing CCA is comparable thus, it is ok.
```
