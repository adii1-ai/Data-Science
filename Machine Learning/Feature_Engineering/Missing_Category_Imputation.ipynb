- This technique is used for categorical data imputation.
- In this data 'missing' word is imputed at the missing place.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gender = np.random.choice(['male', 'female'], size = 20)
city = np.random.choice(['nagar','mandve'], size = 20)
```

```python
gender
```

```python
city
```

```python
data = {
    "Gender" : [None, 'female', 'male', None, 'female', 'female', 'female',
       'female', 'female', 'female', None, 'male', 'male', 'male',
       'male', 'female', 'female', 'female', 'male', None],
    'City' : ['nagar', None, 'mandve', 'mandve', 'nagar', 'nagar', 'mandve',
       'nagar', None, 'nagar', 'mandve', 'nagar', 'mandve', 'nagar',
       'nagar', 'nagar', 'mandve', 'nagar', 'nagar', None]
}
```

```python
df = pd.DataFrame(data)
```

```python
df['Gender'].value_counts().plot(kind= 'kde')
```

```python
df.head()
```

```python
# imputing the 'missing' keyword
```

```python
from sklearn.impute import SimpleImputer
```

```python
df['Gender'].value_counts().plot.bar()
```

```python
df['City'].value_counts().plot.bar()
```

```python
imputer = SimpleImputer(strategy = 'constant', fill_value = 'Missing')
```

```python
df[['Gender']] = imputer.fit_transform(df[['Gender']])
```

```python
df['Gender'].value_counts()
```

```python
df[['City']] = imputer.fit_transform(df[['City']])
```

```python
df['City'].value_counts()
```

```python
df['Gender'].value_counts().plot.bar()
```

```python
df['Gender'].value_counts().plot(kind ='kde')
```
