- It is also called as MICE Imputation.
- MICE - Multivariate Imputer by Chaining Equation
- It is a Iterative process to impute the value.

- The Main thing is --> it only applied on input data.

```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(
    {
        'Age' : [25,30,np.nan,20,59,20,24,np.nan,60,85],
        'Salary' : [30000,28000,20500,36000,50000,20000,49000,70303,90333,86000],
        'Experience':[2,np.nan,4,6,3,5,9,6,1,np.nan]
    }
)
```

```python
df
```

```python
X = df.drop(columns = ['Salary'])
y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.4, random_state = 2)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
mice_imputer = IterativeImputer(max_iter = 10, random_state = 42)

imputed_array_X_train = mice_imputer.fit_transform(X_train)

df_imputed_X_train = pd.DataFrame(imputed_array_X_train,columns = ['Age','Experience'])

imputed_array_X_test = mice_imputer.fit_transform(X_test)
```

```python
df_imputed_X_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2_score(y_pred, y_test)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_pred, y_test)
```
