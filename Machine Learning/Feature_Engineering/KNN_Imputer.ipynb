```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
df = titanic.drop(columns =['sex','sibsp','parch','embarked','class','who','embark_town','alive','alone','adult_male','deck'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = ['survived'])
y = df['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 2)
```

```python
X_train.head()
```

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer(n_neighbors = 5, weights = 'distance') # weights = 'distance'/'uniform'

X_train_trf = knn.fit_transform(X_train)
X_test_trf = knn.transform(X_test)
```

```python
pd.DataFrame(X_train_trf, columns = X_train.columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf, y_train)

y_pred = lr.predict(X_test_trf)

accuracy_score(y_test,y_pred)
```

```python
# Simple Imputer --> mean
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'mean')

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.fit_transform(X_test)
```

```python
lr2 = LogisticRegression()
lr2.fit(X_train_trf2, y_train)
y_pred2 = lr2.predict(X_test_trf2)

accuracy_score(y_pred2, y_test)
```

```python
# It is seen that with simple Imputer accuracy is less and Using KNN Imputer Accuracy increases.
```
