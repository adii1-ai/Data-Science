### Imputinfg the values that are at the extreme ends of distribution of data.


- We can simply use fill() to impute the vlaues.
- End of distribution values are founded as ->

    - EoD = mean  + 3 * Standard Deviation

    - EoD = mean - 3 * Standard Deviation

```python
import numpy as np
import pandas as pd
```

```python
np.random.randint(1,100,20)
```

```python
data = {
    'Age' : [80, np.nan, 69, 24, 95, np.nan, 69, 99, 61,  6,  5, np.nan, 92, 42, 99, 81, np.nan,
       66, 21, np.nan],
    'Salary' : [ np.nan, 1700,  5029, 160, np.nan, 22000,  7000, np.nan, 7001, 40205, 40326, 7300, 40207, 54200, 4800, 82000, 22000,
       31230, 55121,  73232]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df['Age'].plot(kind = 'kde')
```

```python
df['Salary'].plot(kind= 'kde')
```

```python
Age_EOD = df['Age'].mean() + 3 * df['Age'].std()
Age_EOD
```

```python
Salary_EOD = df['Salary'].mean() + 3 * df['Salary'].std()
Salary_EOD
```

```python
df['Age'] = df['Age'].fillna(Age_EOD)
df['Salary'] = df['Salary'].fillna(Salary_EOD)
```

```python
df
```

## Filling Multiple Columns --> using for loop

```python
np.random.randint(10000,100000, 20)
```

```python
data_1 = {
    'Age' : [np.nan, 76, 88,  np.nan, 55, 95,  6, 71,  5, np.nan, 53, 72, 32, 46, 75, 37,  3,
       52, 85, 57],
    'Salary' : [29292, 98426, np.nan, 62023, 78554, np.nan, 53372, 13656, 93602,
       22909, 49035, 25068, 83256, 68070, 21325, 38208, 33796, 42488,
       22602, 17865]
}
```

```python
df_1 = pd.DataFrame(data_1)
```

```python
df_1.head()
```

```python

for cols in df_1:
    EOD = df_1[cols].mean() + 3 * df_1[cols].std()
    df_1[cols] = df_1[cols].fillna(EOD)
```
