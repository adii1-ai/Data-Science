## Another way to access data

### .loc()

```python
# Syntax ->> df.loc[[row lables] , [column lables]] === df.loc[[row_start_lable : row_stop_lable: step], [col_start_lable: col_stop_lable : step]]
```

```python
# It returns the actual value of index/label
# Syntax ->> df.loc[row_label, column_label]

# It pass it the location not the indexes i.e see below line in which we have passed the location 0 to 8 and it will print the data from 0 to 8 th location 
# If this was indexes then data will printed from 0 index to 7 index i.e less one than last index

# Lets see the Difference

x = df.loc[0:8] # location is passed
print(x)  # printed from index 0 to 8

print("\n")

y = df[0:8]
print(y) # printed from index 0 to 7
```

### .iloc()

```python
# Stands for index location
# df.iloc[row_index, column_index] === df.iloc[start_index : stop_index : step , start_index : stop_index : step]
```

```python
df1
```

```python
df1.iloc[2] # it only returns the 2nd row with its column names/lables
```

```python
# To access the particular data from dataFrame use row_index and column_index

df1.iloc[1,2] # Data from 1st row and 2nd column is printed
```

```python
# We can also access the overall Dataframe  as Below ->>

df1.iloc[:,:]
```

```python
df1.iloc[:2] # It returns the only 2 rows i.e rows of index 0 and 1
```

```python
df1.iloc[4:7] # slicing using iloc()
```

```python
df1.iloc[4:7, 2:6] # slicing 4:7 is for rows and slicing 2:6 is for columns
```

```python
df1.iloc[2:5, 0:8] # for slicing we have used the indexes not the location 
```

```python
df1.iloc[4:7, ::2] # if we used double colon then valu after the second colon act as a step means the columns having gap between them is step value are only printed
```

```python
# we can also use this for rows

df1.iloc[::3, 0:8] # rows  after the distance 3 are only printed
```

```python
df1.iloc[2:8:2, 0:8:3] 
```

```python
df1.iloc[2:8:2, 0:8]
```

```python
df1
```

```python
# We cann also access the particular rows and columns by passing the lists of rows and indexes 

df1.iloc[[1,2,3,4, 6,8],[1,3,5]]
```

```python
# Difference between the df.iloc() and df.loc() ->> 
# 1. df.loc() ->> in loc() we can pass the lables i.e column names not the indices
# 2. df.iloc() ->> in iloc() we can pass the indices not the lables 

# see with example

print(df1)

print("\n")
loc = df1.loc[[1,3,4,6,8], ['First Name' ,'Last Name', 'Age']]
print("By using Location : \n\n",loc)

print("\n")
index_loc = df1.iloc[[1,3,4,6],[1,4,6]]
print("By  Using Index Location : \n\n",index_loc)
```

```python
# We can perform the slicing after accessing the particular columns on that columns 

df1[['First Name' ,'Gender', 'Age']].iloc[2:8:2, 0:2]  # while slicing we used the column indices that can be applied only on the accessed columns i.e 0:2 ->> can access only the 0th and 1st index column from ['First Name' ,'Gender', 'Age']
```
