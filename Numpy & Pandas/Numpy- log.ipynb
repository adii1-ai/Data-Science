## Functions

```python
# 1. np.log()  # Natural log
# 2. np.log2()
# 3. np.log10()

# This functions are used to reduce skewness of data
```

### 1. np.log() (Natural Log)

```python
np.log(34)
```

```python
np.log(-10) # we can not find the log of Negative Number, It gives Not a Number
```

```python
np.log(0) # -infinity
```

```python
np.log(1) # 0
```

### 2. np.log2()

```python
np.log2(34)
```

```python
np.log2(10)
```

### 3. np.log10()

```python
np.log10(34)
```

```python
np.log10(10)
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
dict1 = {"Age":np.random.randint(20, 50, size = 10),
         "Height":np.random.randint(160, 180, size = 10),
         "Weight":np.random.randint(60,80, size = 10)
        }
dict1
```

```python
df = pd.DataFrame(dict1)
df
```

```python
np.log(df['Weight']) # this is transformation of data
```

```python
dict2 = {"Age":np.random.randint(20, 50, size = 10),
         "Height":np.random.randint(160, 180, size = 10),
         "Weight":np.random.randint(60,80, size = 10)
        }
dict2
```

```python
df = pd.DataFrame(dict2)
df
```

```python
np.log(df) # This is transformation(log Transformation) , All the values returned are the log of dataframe
```
