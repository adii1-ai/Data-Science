# Numpy - Trigonometry

```python
import numpy as np
```

```python
# sin     >> O/H (Opposite / Hypotenous)
# cos     >> A/H (Adjacent / Hypotenous)
# tan     >> O/A
# cot     >> A/O
# sec     >> H/A
# cosec   >> H/O
```

```python
np.pi # In radian pi = 3.14 and In Degree pi = 180
```

```python
PI = np.pi
Degree = np.rad2deg(PI)
print(Degree)
```

```python
sin30 = np.sin(np.pi/6)  # 180/6 >> 30
print(sin30)
```

```python
sin60 = np.sin(np.pi/3) # 180/3 >> 60
print(sin60)
```

```python
np.cos(np.pi/2) # 180/2 >> 90
```

```python
round(np.cos(np.pi/2)) # It will return zero because e-17 inside a value
```

# Deree to Radian

```python
# It converts the degree into radian
```

```python
arr = np.array([0, 30, 45, 60,90])
arr1 = np.deg2rad(arr)
arr1
```

```python
np.sin(30)  # This is not allowed because it takes argument in radian not in degrees
            # Here 30 is considered as redian thus it does not give expected value of sin 30 degree insteed it gives value of sin 30 radian
```

```python
np.sin(np.pi/6) # This is correct because here argument passed is in radian
```
