## 1. np.mean()

```python
import numpy as np
arr = np.random.randint(160, 180, size=20)
print(arr)

Mean = np.mean(arr)
print("Mean is : ",Mean)
```

## 2. np.median()

```python
# Median is the middle value from the array.
# If the length of an array is odd then median is the exact middle value present in the sorted array.
# If the length of an array is even then median is the mean of two middle values.
# for finding the Median first sort the array.
```

```python
arr = np.array([30,23,50,76,55,79,66]) # odd length 
print("Array : ", arr)
sorted_arr = np.sort(arr)
print("Sorted Array : ", sorted_arr)
Median = np.median(sorted_arr)
print("Median is : ", Median)
```

```python
arr = np.array([30,23,50,76,80,55,79,66]) # even length
print("Array : ", arr)
sorted_arr = np.sort(arr)
print("Sorted Array : ", sorted_arr)
Median = np.median(sorted_arr)
print("Median is : ", Median) # (55+66)/2 i.e Median = mean of middle two element
```

## 3. np.std() (standard Deviation)

```python
# deviation means the scatterdness(distribution) around the central tendency(mean)
```

```python
arr = np.array([5,4,1,8,9,3])
np.mean(arr)
```

```python
np.std(arr)
```

```python
arr = np.random.randint(10, 50, size = 20)
print("Array is : ",arr)
mean = np.mean(arr)
print("Mean : ", mean)
Standard_Deviation = np.std(arr)
print("Standard Deviation is : ",Standard_Deviation)
```

```python
arr = np.random.randint(10, 30, size = 10)
print("Array is : ",arr)
mean = np.mean(arr)
print("Mean : ", mean)
Standard_Deviation = np.std(arr)
print("Standard Deviation is : ",Standard_Deviation)
```

```python
18.9 + 6.41  # positive standard deviation
```

```python
18.9 - 6.41  # Negative standard deviation
```

```python
# If we draw the curve of above array then we got the standard deviation about the mean value
# positive std ->> mean + std  (This is Maximun value Distributed around the mean)
# Negative std ->> mean - std  (This is Maximun value Distributed around the mean)
```

## 4. np.var() (Variance)

```python
# Variance is the square of standard Deviation
```

```python
arr = np.random.randint(10, 30, size = 10)
print("Array is : ",arr)
mean = np.mean(arr)
print("Mean : ", mean)
Standard_Deviation = np.std(arr)
print("Standard Deviation is : ",Standard_Deviation)
var = np.var(arr)
print("Variance is : ",var) # square of std
```

## 7. mode()

```python
# Mode is the most repeated values or the most occcured value in the array
# for findig mode we have to import mode from scipy library or we can also use dataframe (df) i.e df.mode()
```

```python
arr = np.random.randint(40,60,size = 10)
arr
```

```python
from scipy.stats import mode
Mode = mode(arr)
print(Mode)
print(type(Mode))
```

```python
Mode[0]  # It gives which value iss most repeated
```

```python
Mode[1]  # It gives the how many times the value is repeated 
```

```python
# by using dataframe

import pandas as pd
df = pd.DataFrame(arr)
print(df)
df.mode
```
