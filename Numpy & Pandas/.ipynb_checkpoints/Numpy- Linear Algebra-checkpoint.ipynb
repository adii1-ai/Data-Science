# Numpy - Linear Algebra

```python
import numpy as np
```

```python
2x + 5y = 15
3x + 4y = 20

x = ?
y = ?
```

```python
arr_A = np.array([[2,5],[3,4]])
arr_B = np.array([15, 20])

print("Array A is : \n", arr_A)
print("Array B is : \n", arr_B)

result = np.linalg.solve(arr_A,arr_B)
print("Result : ",result)
```

```python
3x + 5y + 6z = 30
4x + 7y + 8z = 20
4x + y + 9z = 43

x = ?
y = ?
z = ?
```

```python
arr_A = np.array([[3,5,6],[4,7,8],[4,1,9]])
arr_B = np.array([30,20,43])

print("Array A is : \n", arr_A)
print("Array B is : \n", arr_B)

Result = np.linalg.solve(arr_A, arr_B)
print("Result : ", Result)
```
