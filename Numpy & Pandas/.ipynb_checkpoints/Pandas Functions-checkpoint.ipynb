# Pandas Functions

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('test.csv')
df
```

## 1. unique()

```python
# It returns all the unique values present inside the particular column

df['Country'].unique()
```

```python
df['First Name'].unique()
```

## 2. nunique()

```python
# it used to return the number of unique values
```

```python
df['First Name'].nunique() 
```

```python
df['Age'].nunique() 
```

## 3. value_counts()

```python
# It returns the count of each unique value present in the particulat column
```

```python
df.iloc[1,5] = 32
```

```python
df
```

```python
df['Age'].value_counts()
```

### Finding Percantage of each Unique value

```python
# To calculate percantage of each unique value pass paramater normalize = True to the value_count()
```

```python
df['Age'].value_counts(normalize = True)
```

```python
df['First Name'].value_counts(normalize = True)
```

## 4. .to_dict() 

```python
# There are 2 ways to convert the above into dictionary
```

```python
# 1st way

df['Age'].value_counts(normalize = True).to_dict()
```

```python
# 2nd way

dict(df['Age'].value_counts(normalize = True))
```

## 5. crosstab()

```python
# the following represents the count of each unique name with gender wise 
# i.e the hown many males and females having the same name

df2 = pd.crosstab(df['First Name'], df['Gender'])
df2
```
