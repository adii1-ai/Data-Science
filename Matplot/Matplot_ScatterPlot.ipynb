# Scatter Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## plt.scatter(x,y)

```python
x = np.linspace(-10,10,50)
y = 10*x+3+np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
df = pd.read_csv('batter.csv')
df
```

```python
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'], color = 'blue', marker='D')
plt.title('Avg and Strike rate Analysis of top 50 Batsman')
plt.xlabel('Average per Match')
plt.ylabel('Strike Rate')
```

```python
tips = sns.load_dataset('tips')
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*20)   # size represents the number of customers come together 
```

```python
# Scatter Plot using 2D Plot

plt.plot(tips['total_bill'], tips['tip'],'o')
```
