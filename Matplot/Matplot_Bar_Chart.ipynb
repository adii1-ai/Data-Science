# Bar Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
children = [10,20,40,30,50]
colors = ['Red','Green','Blue','Yellow','Pink']
plt.bar(colors,children)
```

## Horizontal Bar Graph

```python
children = [10,20,40,30,50]
colors = ['Red','Green','Blue','Yellow','Pink']
plt.barh(colors,children, color='Black')
plt.title('Colors vs Number of Childrens Analysis')
plt.xlabel('Number of Childrens')
plt.ylabel('Color')
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],label='2016')
plt.bar(df['batsman'],df['2017'],label='2017')

plt.legend()
```

```python
# To print the side by side bars following code is used.

plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width = 0.2,color = "blue")
plt.bar(np.arange(df.shape[0]),df['2016'],width = 0.2, color = "Green")
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width = 0.2, color = "Red")

plt.show()
```

## plt.xticks(existing_axis_name, new_name_to_replace)

```python
# plt.xticks() is used to replace the existing name with the new name.
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width = 0.2,color = "blue")
plt.bar(np.arange(df.shape[0]),df['2016'],width = 0.2, color = "Green")
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width = 0.2, color = "Red")

plt.xticks(np.arange(df.shape[0]), df['batsman']) # This is to replace the numbers on x axis with the batsman name

plt.title("Batsman and there Runs Analysis")

plt.xlabel("Batsman")
plt.ylabel("Runs")

plt.show()
```

### xticks() parameters >>

## 1. xticks(rotation = "vertical") 

```python
# Used when the names on x-axis are too long and overlap on each others.
# It places the names verticlly
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width = 0.2,color = "blue")
plt.bar(np.arange(df.shape[0]),df['2016'],width = 0.2, color = "Green")
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width = 0.2, color = "Red")

plt.xticks(np.arange(df.shape[0]), df['batsman']) # This is to replace the numbers on x axis with the batsman name

plt.title("Batsman and there Runs Analysis")

plt.xlabel("Batsman")
plt.ylabel("Runs")

plt.xticks(rotation = "vertical")

plt.show()
```
