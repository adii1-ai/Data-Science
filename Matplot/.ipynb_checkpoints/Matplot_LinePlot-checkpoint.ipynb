# Line Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ploting the line chart or graph :
# Syntax ->> plt.plot(x,y)

price = [48000,38000,49000,58000,50000]
year = [2021,2022,2023,2024,2025]
plt.plot(year, price)
```

```python
# Dataset used 

batsman = pd.read_csv("sharma-kohli.csv")
batsman
```

```python
# Plotting the graph of virat kohli with the runs scored in particular year

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# Plotting the graph of Rohit Sharma with the runs scored in particular year


plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# Plotting the common graph for both Virat and Rohit

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
```

## plt.xlabel() and plt.ylabel()

```python
# we can give the lables/names to the x-axis and y-axis : 
# Syntax ->>  plt.xlabel("label")
#             plt.ylabel("label")

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

## Parameters to the plt.plot()

### 1. color

```python
# Matplot also give flexibilty to cchange the  color of the graph or the lines of a graph

plt.plot(batsman['index'],batsman['V Kohli'], color="green")
plt.plot(batsman['index'],batsman['RG Sharma'], color="red")
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

### 2. linestyle

```python
# we can also change the style of a line by using parameter linestyle = 'style_name'

# Line Styles ->>
# 1. solid line
# 2. dashed line
# 3. dotted line
# 4. dashdot

```

```python
# lnestyle = solid (default)

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# lnestyle = dotted

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'dotted')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'dotted')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# lnestyle = dashed

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'dashed')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'dashed')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# lnestyle = dashdot

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'dashdot')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# We can use defferent linestyle for different lines

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'dashed')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

### 3. linewidth

```python
# We can also change the width of the line by passing parameter linewidth 

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=2)
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=5)
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

## 4. Marker

```python
# marker = 'D' ->> It prints the Diamond on on the position of data point 

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=3,marker='D')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# marker = '.' ->> It prints the point or dot on on the position of data point 

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='.')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=1,marker='.')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# marker = '+' ->> It prints the + sign on on the position of data point 

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='+')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=1,marker='+')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# marker = '*' ->> It prints the Asteric(*) on on the position of data point 

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='*')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=1,marker='*')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# marker = 'o' ->> It prints the Big Dot pr Circle on on the position of data point 

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='o')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=3,marker='o')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
# marker = '>' ->> It prints the Forward Arrow on on the position of data point 
# marker = '<' ->> It prints the Backward Arrow on on the position of data point 

plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='<')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=1, marker='>')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

## 5. markersize

```python
# This parameter  is used to conrol the size of marker 
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='o', markersize=3)
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=1, marker='o',markersize=7)
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='D',markersize=5)
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=3,marker='D',markersize=10)
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')
```

## 6. label

```python
plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='D',markersize=5, label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=3,marker='D',markersize=10, label='Rohit')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')

plt.legend(loc='upper right')               # to print the which line represents which data (Upper Left Corner Box)
```

```python
plt.plot(batsman['index'],batsman['V Kohli'], color="green", linestyle = 'solid', linewidth=1, marker='D',markersize=5, label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color="red",linestyle = 'solid',linewidth=3,marker='D',markersize=10, label='Rohit')
plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel('Year')
plt.ylabel('Runs Scored')

plt.legend(loc='center left')
```

# Triming of Data

```python
price = [48000, 57000,49000, 47000,45000, 4500000]
year = [2020,2021,2022,2023,2024,2025]
plt.plot(year, price)
```

```python
price = [48000, 57000,49000, 47000,45000, 4500000]
year = [2020,2021,2022,2023,2024,2025]
plt.plot(year, price)
plt.ylim(0, 100000)
plt.xlim(2020, 2024)
```

## plt.grid()

```python
price = [48000, 57000,49000, 47000,45000, 4500000]
year = [2020,2021,2022,2023,2024,2025]
plt.plot(year, price)
plt.ylim(0, 100000)
plt.xlim(2020, 2024)
plt.grid()
```
