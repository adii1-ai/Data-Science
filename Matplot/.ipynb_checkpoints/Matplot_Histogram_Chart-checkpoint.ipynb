# Histogram Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [32,45,56,10,15,27,64]
plt.hist(data)
```

# Parameters

# 1. bins=[]

```python
data = [32,45,56,10,15,27,64]
plt.hist(data, bins=[10,15,20,25,30,35,40,45,50,55,60,65,70])
```

## Logarithemic scale

```python
arr = np.load('big-array.npy')
arr
```

```python
arr.shape
```

```python
# Without log

plt.hist(arr, bins = [10,20,30,40,50,60,70,80,90])
plt.show()
```

```python
# With log


plt.hist(arr, bins = [10,20,30,40,50,60,70,80,90], log = True)
plt.show()
```
