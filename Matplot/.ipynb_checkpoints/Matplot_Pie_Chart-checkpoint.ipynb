# Pie Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [23,45,100,20,49]
plt.pie(data)
plt.show()
```

```python
data = [23,45,100,20,49]
subjects = ['English', 'Science', 'Math', 'Geography', 'Biology']
plt.pie(data, labels = subjects)
plt.show()
```

## Example

```python
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'])
plt.show()
```

## parameters

### 1. autopct = '%0.1f%%'

```python
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%')
plt.show()
```

### 2. colors = []

```python
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', colors=['blue', 'red', 'yellow', 'black', 'green','maroon'])
plt.show()
```

### 3. explode = []

```python
# Inside the array we can pass the values that represents how much we want to explode the particular section of a pie chart
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%',explode = [0.3,0,0,0,0,0.1])
plt.show()
```

### 4. shadow = True

```python
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%',explode = [0.3,0,0,0,0,0.1], shadow = True)
plt.show()
```
