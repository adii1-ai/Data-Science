```python
iris.sample(5)
```

```python
# Replacing the spacies from Spcies column

# function ->>   colummn_name.replace({})

iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'])
plt.xlabel('Sepal length')
plt.ylabel('Patal length')
plt.title('Iris Flower Analysis')
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'magma')
plt.xlabel('Sepal length')
plt.ylabel('Patal length')
plt.title('Iris Flower Analysis')
```

## Annotation

### plt.text(point_coordinates, 'name_to_point')

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)

plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Ponit 3')
plt.text(4,8,'Point 4')
```

### parameters to plt.text()

### 1. fontdict={'size':    , 'color':'    '}

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)

plt.text(1,5,'Point 1', fontdict={'size':20, 'color':'brown'})
plt.text(2,6,'Point 2', fontdict={'size':12, 'color':'blue'})
plt.text(3,7,'Ponit 3', fontdict={'size':10, 'color':'red'})
plt.text(4,8,'Point 4', fontdict={'size':15, 'color':'green'})
```

```python
batters = pd.read_csv('batter.csv')
batters
```

```python
sample_df = batters.head(100).sample(25,random_state = 5)
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

```python
plt.figure(figsize = (15,10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# To represent the points size based on runs scored pass the parameter ->>  s = sample_df['runs']

plt.figure(figsize = (15,10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'])
            
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])  # -->> plt.text(x,y, 'point_name')
```

# Horizontal and Verticle lines

```python
plt.figure(figsize = (18,10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'])

plt.axhline(130, color = 'red') # Horizontal line
plt.axvline(30, color = 'red') # Verticle Line
            
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

## Two Ways to plot Graph

### First Way

```python
 # To give the size to a Graph
plt.figure(figsize = (15,7))

# To plot graph
plt.scatter(batters['avg'], batters['strike_rate'])

 # To Set name of Graph
plt.title('Avg VS Strike Rate')

 # To give axis Names/labels
plt.xlabel('Average')
plt.ylabel('Strike Rate')

 # To sho the Graph
plt.show()
```

### Second way

```python
 # plt.subplot() takes 2 objects i.e figure and axes Which is then stored inside a variables fig and ax.
fig, ax = plt.subplots(figsize=(15,7))

 # To plot graph
ax.scatter(batters['avg'], batters['strike_rate'], color='red', marker='D', s = 100)

 # To Set name of Graph
ax.set_title('Avg VS Strike Rate')

 # To give axis Names/labels
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

 # to show the graph
#fig.show()            # In jupyter Notebook this not works We have to write plt.show()

plt.show()
```

## Ploting the sub graph

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex  = True, figsize =(17,15))
# sharex = True ==> shares the x axis values  or the x coordinates

ax[0].scatter(batters["avg"], batters["strike_rate"], color='green', marker = 'D', s = 70)
ax[1].scatter(batters["avg"], batters["runs"],color='red', marker = 'D', s = 70)

ax[0].set_title("Avg Vs Strike Rate")
ax[0].set_ylabel('Strike rate')

ax[1].set_title("Avg Vs Runs")
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

## Plotting the different graphs in subgraphs

```python
# Used when analysing multiple things at once
```

```python
# THIS IS THE CLEANER OPTION TO PLOT THE SUBGRAPHS

fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,10))
# sharex = True ==> shares the x axis values  or the x coordinates

# Above 2 are Scatter Plot
ax[0, 0].scatter(batters["avg"], batters["strike_rate"], color='green', marker = 'D', s = 70)
ax[0, 1].scatter(batters["avg"], batters["runs"],color='red', marker = 'D', s = 70)
ax[1, 0].hist(batters["avg"])
ax[1, 1].hist(batters["runs"])


```

## Another way to plot subplots

```python
fig = plt.figure()

# First Graph
ax1 = fig.add_subplot(2,2,1) # (no. of rows, no.of columns, 1st graph)
ax1.scatter(batters["avg"], batters["strike_rate"], color = 'red')

# Second Graph
ax2 = fig.add_subplot(2,2,2) # (no. of rows, no.of columns, 2st graph)
ax2.scatter(batters["avg"], batters["runs"], color = 'green')

# Third Graph
ax2 = fig.add_subplot(2,2,3) # (no. of rows, no.of columns, 2st graph)
ax2.hist(batters["avg"], color = 'black')

# Fourth Graph
ax2 = fig.add_subplot(2,2,4) # (no. of rows, no.of columns, 2st graph)
ax2.hist(batters["runs"], color = 'blue')
```

# Surface plot

```python
x = np.linspace(-10,10, 100)
y = np.linspace(-10,10,100)

xx, yy, = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize = (12,8))
ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig= plt.figure()
ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.sin(xx)

fig = plt.figure()
ax = plt.subplot(projection = '3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

# Contour Plot

## contour()

```python
#  Contour plot is used to represent the 3D graph into 2D graph.
#  It is 2D plot
# used in machine learning loss function study
```

```python
z = xx**2 + yy**2

fig = plt.figure(figsize = (10,5))
ax = plt.subplot()

p = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
a = np.linspace(-10,10, 50)
b = np.linspace(-10,10,50)

aa, bb, = np.meshgrid(x,y)

xx,yy = np.meshgrid(x,y)
z = aa**3 + bb**3

fig = plt.figure()
ax = plt.subplot()


a = ax.contour(xx, yy, z, cmap = 'viridis')
fig.colorbar(a)
```

## contourf()

```python
# contourf() is used to print the filled contour graph
# used in machine learning loss function study
```

# HeatMap

```python
# Loading Dataset

delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery
```

```python
# Filtering the dataset 

temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
temp_df
```

```python
# creating pivote table containing the count of six in each over
grid = temp_df.pivot_table(index = 'overs', columns = 'ballnumber', values = 'batsman_run', aggfunc = 'count')
grid
```

```python
# PLotting the graph

plt.figure(figsize = (20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21))) # To start the over number from 1 insteed from zero
# in above line basically, overs number is from 0 to 19 which is replaced as, 1 to 20

plt.xticks(np.arange(0,6), list(range(1,7)))
# in above line the ballnumber is from 0 to 5 which is replaced as  1 to 6

plt.colorbar()
```

```python
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df.head()
```
